```python
index_list=[]
index_list.append(list(new_df[new_df.duplicated()].index))
```

```python
for i in index_list:
  data=new_df.drop(new_df.index[i])
  
  
```

```python
data.head()
```

```python
#Feature Engineering
```

```python
#Step 1-> Imputation-> checking missing values
data.isnull().sum()
```

```python
#Step->2 Remove outliers
int_col=[]
for col in list(data.columns):
    if(data[col].dtype=='int64'):
      int_col.append(col)
int_col      
```

```python
new_df=data[['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']]
new_df.head()
```

```python
plt.figure(figsize=(15, 15))
plt.suptitle('Univariate Analysis of Numerical Features Before Outliers Removal', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(0, len(int_col)):
    plt.subplot(5, 3, i+1)
    sns.histplot(x=new_df[int_col[i]],kde=True, color='r')
    plt.xlabel(int_col[i])
    plt.tight_layout()
```

```python
#Capping the outlier rows with Percentiles
for index in int_col:
  upper_lim = new_df[index].quantile(.95)
  lower_lim = new_df[index].quantile(.05)
  new_df.loc[(new_df[index] > upper_lim),index] = upper_lim
  new_df.loc[(new_df[index] < lower_lim),index] = lower_lim
```

```python
plt.figure(figsize=(15, 15))
plt.suptitle('Univariate Analysis of Numerical Features After Outliers Removal', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(0, len(int_col)):
    plt.subplot(5, 3, i+1)
    sns.histplot(x=new_df[int_col[i]],kde=True, color='r')
    plt.xlabel(int_col[i])
    plt.tight_layout()
```

```python
cat_col=[]
for col in list(data.columns):
    if(data[col].dtype!='int64'):
      cat_col.append(col)    
```

```python
cat_data=data[cat_col]
cat_data.head()
```

```python
for index in cat_col:
  print(cat_data[index].value_counts())
  print("=================================")
```

```python
cat_data['workclass']=cat_data['workclass'].str.replace('?','Private')
cat_data['occupation']=cat_data['occupation'].str.replace('?','Other-service')
cat_data['native-country']=cat_data['native-country'].str.replace('?','United-States')
```

```python
check_col=['workclass','occupation','native-country']
for index in check_col:
  print(cat_data[index].value_counts())
  print("=======================================")
```

```python
cat_data.isnull().sum()
```

```python
for index in cat_col:
  encoded_columns=pd.get_dummies(cat_data[index])
  cat_data=cat_data.join(encoded_columns).drop(index,axis=1)
```
